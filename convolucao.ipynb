# Importação das bibliotecas

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
```

# Imagens originais

```python
# Le e mostra a imagem
def mostra_imagem(imagem):
  imagem = cv2.imread(imagem)
  imagem = cv2.cvtColor(src=imagem, code=cv2.COLOR_BGR2GRAY)
  plt.figure(figsize=[8,8])
  plt.imshow(abs(imagem), cmap='gray')
  return imagem
```

# Função de convolução

```python
def convolucao(imagem, kernel):
  # Transforma a imagem para escala cinza
  imagem_cinza = cv2.imread(imagem)
  imagem_cinza = cv2.cvtColor(src=imagem_cinza, code=cv2.COLOR_BGR2GRAY)

  # Faz a convolução
  imagem_convoluida = convolve2d(imagem_cinza, kernel)

  # Mostra a imagem
  # plt.figure(figsize=[8, 8])
  plt.imshow(abs(imagem_convoluida), cmap='gray')

  return imagem_convoluida
```
