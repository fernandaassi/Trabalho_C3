# Filtros

## Média

```python
media = np.array([[1/9, 1/9, 1/9], 
                  [1/9, 1/9, 1/9], 
                  [1/9, 1/9, 1/9]])

convolucao(imagem, media)
```

## Gaussiano

```python
gaussiano = np.array([[1/16, 1/8, 1/16], 
                      [1/8, 1/4, 1/8], 
                      [1/16, 1/8, 1/16]])

convolucao(imagem, gaussiano)
```

## Sobel

```python
# Filtro de Sobel horizontal
sobel_h = np.array([[-1, 0, 1], 
                    [-2, 0, 2], 
                    [-1, 0, 1]])

convolucao(imagem, sobel_h)
```

```python
# Filtro de Sobel vertical
sobel_v = np.array([[1, 2, 1], 
                    [0, 0, 0], 
                    [-1, -2, -1]])

convolucao(imagem, sobel_v)
```

## Prewitt

```python
# Filtro de Prewitt horizontal
prewitt_h = np.array([[-1, 0, 1],
                      [-1, 0, 1], 
                      [-1, 0, 1]])

convolucao(imagem, prewitt_h)
```

```python
# Filtro de Prewitt vertical
prewitt_v = np.array([[-1, -2, -1], 
                      [0, 0, 0], 
                      [1, 1, 1]])

convolucao(imagem, prewitt_v)
```
